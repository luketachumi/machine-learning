```python
import os
```

```python
import pandas as pd
```

```python
ds1 = pd.read_csv('train.csv')
```

```python
ds1.head()
```

```python
ds1.drop(['PassengerId','Name','Ticket','Cabin','Embarked','SibSp','Parch'], axis=1, inplace=True )
```

```python
ds1.isnull().sum()
```

```python
ds1['Age'].describe()
```

```python
ds1['Age'].fillna(ds1['Age'].mean(),inplace=True)
```

```python
ds1.isnull().sum()
```

```python
l_sex_dummies=pd.get_dummies(ds1['Sex'],drop_first=True)
```

```python
ds1= pd.concat([ds1,l_sex_dummies],axis=1)
```

```python
ds1.head()
```

```python
ds1.drop(['Sex'], axis=1, inplace=True )
```

```python
ds1.head()
```

```python
ds1.head()
```

```python
X=ds1.drop(['Survived'],axis=1)
y=ds1['Survived']
```

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC( C= 100,kernel='rbf')
```

```python
model_svc.fit(X, y)
```

```python

df2 = pd.read_csv('test.csv')
```

```python
df2.head()
```

```python
df3=df2.drop(['PassengerId','Name','Ticket','Cabin','Embarked','SibSp','Parch'], axis=1 )
```

```python
df3.isnull().sum()
```

```python
df3['Age'].fillna(df3['Age'].mean(),inplace=True)
df3['Fare'].fillna(df3['Fare'].mean(),inplace=True)
```

```python
l_sex_dummies=pd.get_dummies(df3['Sex'],drop_first=True)
df3= pd.concat([df3,l_sex_dummies],axis=1)
df3.drop(['Sex'], axis=1, inplace=True )
```

```python
df3.head()
```

```python
df3.head()
```

```python
y_predicted = model_svc.predict(df3)
```

```python
submission = pd.DataFrame({
        "PassengerId": df2['PassengerId'],
        "Survived": y_predicted
    })
```
